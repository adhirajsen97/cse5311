```python
import platform
import psutil
import pandas as pd
import matplotlib.pyplot as plt
```

```python
column_names = ['ArraySize', 'BubbleSort', 'InsertionSort', 'SelectionSort']
df = pd.read_csv('sortingAlgs_Benchmark_results.csv', names=column_names)
df.head()
```

```python
# Basic system information
system_info = platform.uname()
print(f"System: {system_info.system}")
print(f"Node Name: {system_info.node}")
print(f"Machine: {system_info.machine}")
print(f"Processor: {system_info.processor}")

# CPU information
cpu_info = platform.processor()
cpu_count = psutil.cpu_count(logical=False)
print(f"CPU: {cpu_info}")
print(f"Physical CPU cores: {cpu_count}")

# Memory information
memory_info = psutil.virtual_memory()
print(f"Total memory: {memory_info.total / (1024**3):.2f} GB")

# Disk information
disk_info = psutil.disk_usage('/')
print(f"Total disk space: {disk_info.total / (1024**3):.2f} GB")
```

```python
# Plotting
plt.figure(figsize=(10, 6))

plt.plot(df['ArraySize'], df['BubbleSort'], label='Bubble Sort')
plt.plot(df['ArraySize'], df['SelectionSort'], label='Selection Sort')
plt.plot(df['ArraySize'], df['InsertionSort'], label='Insertion Sort')

plt.title('Sorting Performance')
plt.xlabel('Array Size')
plt.ylabel('Time Taken (ms)')
plt.legend()
plt.grid(True)

plt.show()
```
