```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the Excel file into a DataFrame
df = pd.read_excel('quick_sort_times.xlsx')

# Extract data for plotting
n = df['n']
best_time = df['Best Time']
average_time = df['Average Time']
worst_time = df['Worst Time']

# Plot the graph
plt.figure(figsize=(10, 6))
plt.plot(n, best_time, label='Best Time', marker='o')
plt.plot(n, average_time, label='Average Time', marker='s')
plt.plot(n, worst_time, label='Worst Time', marker='^')
plt.title('Quick Sort Times')
plt.xlabel('n (Size of Input Array)')
plt.ylabel('Time (nanoseconds)')
plt.grid(True)
plt.legend()
plt.xscale('log')  # Log scale for better visualization
plt.yscale('log')  # Log scale for better visualization
plt.tight_layout()
plt.show()
```
